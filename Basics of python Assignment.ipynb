```python
# 1.Program to find a Prime Number or Not
num = int(input("Enter a number: "))

if num > 1:
    for i in range(2, (num//2)+1):
        if num % i == 0:
            print(num, "is not a Prime number")
            break
    else:
        print(num, "is a Prime number")
else:
    print(num, "is not a Prime number")
```

```python
# 2.Product of Random Numbers
import random

#creating random numbers
number1=random.randint(1,10)
number2=random.randint(1,10)

print("First Number: ",number1)
print("Second Number: ",number2)

UserAns=int(input("Enter the product of these two Numbers: "))

#Test if it matches or not
if(UserAns == number1*number2):
    print("Correct Answer!")
else:
    print("Wrong Answer")
    print("Correct Product is:",number1*number2)
```

```python
# 3.Squares of Even Numbers
print("Squares of even numbers between 100 and 200:")

for i in range(100, 201):
    if i % 2 == 0:   
        print(i, ":", i * i)
```

```python
# 4.Hospital Billing System

Room_charge_perday = 2000
Doctor_fee = 1500
lab_test_fee = 300
medicine_charge = 2400
day_stayed = 4
lab_tests_taken = 3

#Calculations
room_charge = Room_charge_perday * day_stayed
lab_charge = lab_test_fee * lab_tests_taken

Total_Bill = medicine_charge + Doctor_fee + room_charge + lab_charge

# Apply if discount is applicable
if( Total_Bill > 10000):
    discount = Total_Bill * 0.10
else:
    discount=0

print("Room Charge: ",room_charge)
print("Lab Charge: ",lab_charge)
print("Total Bill: ",Total_Bill)
print("Discount: ",discount)
print("Final Bill: ",Total_Bill - discount)
```
