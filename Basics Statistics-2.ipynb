```python
import pandas as pd

#Load tha data set
df1 = pd.read_csv("Patient_Data.csv")
df2 = pd.read_csv("Billing_Data.csv")
#displays data type information for every variable
df1.info()
```

```python
#Droping the unwanted columns from the data while printing(Not permanently)
df1 = df1.drop(df1.columns[[5,6]],axis=1)
df1
```

```python
#Selecting only billing revelant columns by neglecting Name
df1=df1[['PatientID', 'Department', 'Doctor', 'BillAmount']]
df1
```

```python
# Total BillAmount per department(duplicates included)
Total_BillAmount = df1.groupby('Department')['BillAmount'].sum()
Total_BillAmount
```

```python
#Dropping duplicates records
df1 = df1.drop_duplicates(subset='PatientID')
df1
```

```python
import numpy as np

#Finding mean value for bill amount to fill missing value with 2decimal places
Mean = np.round(df1['BillAmount'].mean(),2)
print(Mean)
```

```python
#Fill the blank with mean value
df1["BillAmount"].fillna(value=Mean,inplace=True)
df1
```

```python
#Merging the billing dataset with patient dataset on PatientID by using concat.
merged_df = pd.merge(df1,df2, on='PatientID')
merged_df
```

```python
#Creating dataFrame for new patients(row-wise concat)
new_patients_row = pd.DataFrame({
    'PatientID': [106, 107],
    'Department': ['Cardiology', 'Orthopedics'],
    'Doctor': ['Dr.Lee', 'Dr Mehta'],
    'BillAmount': [7000, 8500],
})

#creating dataFrame for new patients(Column-wise concat)
new_billing_columns = pd.DataFrame({
    'InsuranceCovered': [2500,3000],
    'FinalAmount': [7000,8500]
})
```

```python
#Concatenate new patients (concadinating column wise)
result_data_set = pd.concat([new_patients_row,new_billing_columns],axis=1)
result_data_set 
```

```python
# Concatenates merged and new patient data row-wise, resets index, and starts indexing from 1
Final_dataset = pd.concat([merged_df,result_data_set],ignore_index=True)
Final_dataset.index+=1

#This is the final cleaned data set
Final_dataset
```
