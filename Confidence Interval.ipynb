```python
........Confidence Interval:...........
A confidence interval (CI) is a range of values calculated from sample data that is likely to contain the true population parameter
with a certain level of confidence.
```

```python
import numpy as np
from scipy import stats

# Given data
data = np.array([
    1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85,
    1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29
])

# Common calculations
n = len(data)
mean = np.mean(data)

print("Sample size (n):", n)
print("Sample mean:", round(mean, 4))
```

```python
.......A. 99% Confidence Interval Using Sample Standard Deviation:........
```

```python
# sample standard deviation
sample_std = np.round(data.std(),2)
print("Sample Standard eviation:",sample_std)
```

```python
Step 1---->Identify the appropriate distribution:

    We use the t-distribution because:
         *The population standard deviation is unknown.
         *The sample size is small (n<30).
   Degrees of freedom:
        df = n-1
```

```python
df = n-1
print("Degree of freedom:",df)
```

```python
#Step 2-----> Find the critical value:

from scipy.stats import t

#confidene level 99%
confidence_level = 0.99

#alpha value
alpha = 1 - confidence_level

# Calculates the two-tailed t critical value for the given confidence level and degrees of freedom
t_critical = np.round(t.ppf(1 - alpha/2, df),4)
print("Critical point:",t_critical )
```

```python
#Step 3----> Compute the margin of error:

# Computes the margin of error for a confidence interval using the t-distribution
margin_error = np.round( t_critical * (sample_std / np.sqrt(n)),4)
print("Margin Error:",margin_error) 
```

```python
#Step 4:finding confidence interval

#Formula
Condfidence_interval = np.round((mean - margin_error ,mean + margin_error),4)
print("Confidence Interval 99% lies between:",Condfidence_interval)
```

```python
Conclusion:

          We are 99% confident that the true mean durability of the print-heads lies between 1.09 and 1.39 million characters.
```

```python
.........B. 99% Confidence Interval Using Known Population Standard Deviation:...........
```

```python
Population_std_deviation = 0.2
```

```python
Step 1---->Choose the distribution

      We use the z-distribution because:
              *The population standard deviation is known
              *The sampling distribution of the mean is normal
```

```python
#Step 2----->Find the critical z-value

from scipy.stats import norm

#99% confidence level
confidence_level = 0.99
alpha = 1 - confidence_level

# Calculates the two-tailed Z critical value for a 99% confidence level
z_critical = np.round(norm.ppf(1 - alpha/2),4)
print("z critical value:",z_critical)
   
```

```python
#Step 3----> Compute the margin of error:

margin_error = np.round(z_critical * (Population_std_deviation / np.sqrt(n)),4)
print("Margin Error:",margin_error)
```

```python
#Step 4----> 99% Confidence interval:

Confidence_interval = np.round((mean - margin_error, mean + margin_error),4)
print("Confidence interval",Confidence_interval)
```

```python
Conclusion:
        We are 99% confident that the true mean durability lies between 1.11 and 1.37 million characters, assuming the population standard 
deviation is exactly 0.2 million characters.
```
