```python
Data provided:
  1. Weekly cost model: 
       W = $1,000 + $5X

  2. Mean units produced per week(Entire data mean):
       μ = 600

  3. Standard deviation of units produced:
       σ = 25

  4. Sample size:
       n = 25

  5. Sample mean weekly cost:
       xˉ = 3050

  6. Significance level(Alpha):
       α = 0.05
```

```python
Hypothesis Testing:
  *Hypothesis testing is a statistical method used to make a decision about a population based on sample data.
  *Hypothesis testing is a process to decide whether an assumption about data is correct, based on sample evidence.

      The restaurant owners claim that the weekly operating cost is higher than the model suggests.

  1. Null Hypothesis-H0
       The null hypothesis (H0) states that there is no change, no effect or no difference in the situation being studied.
            
            μ = 4000
          
  2. Alternative Hypothesis
      The alternative hypothesis (H1) states that there is a change, effect or difference from the assumed condition.
      
            μ > 4000

  This is a right-tailed test
```

```python
2. Calculating the Test Statistic:

    Test Statistic Formula:
                 z = (Sample_mean - Population_Mean) / (Standard_deviation / math.sqrt(n))

Note:
   To make ztest & p_value, sample data should be in array or list form.But data is in single digit.So we calculate ztest by manually by using formula
instead of inbuild functions
```

```python
#Substituting values

import math

Sample_mean =3050
Population_Mean = 4000 
Standard_deviation =125
n = 25

z_test = (Sample_mean - Population_Mean) / (Standard_deviation / math.sqrt(n))
print("z-test value:",z_test)
```

```python
3. Comparing with alpha value

  Critical value for 5% = 0.05 (right-tailed test)
                Critical_val = 1.645 (positive because it is right-tailed test)
                            
 case 1: If ztest value falls above critical value---> H0 is accepted and H1 is rejected
 case 2: If ztest value falls under critical value---> H1 is accepted and H0 is rejected
```

```python
if z_test > 1.645:
    print("H0 is accepted and H1 is rejected")
else:
    print("H1 is accepted and H0 is rejected")
```

```python
4.Decision making:
 * We obtained a ztest value as -38.0 and critical value as 1.645 , ztest value is lesser than critical value.
 * So Null hypothesis is rejected and Alternative value is accepted based on statistical significance measures.
```

```python
5.Conclusion:
     There is strong evidence that the weekly operating costs are higher than the theoretical model suggests.By analyzing ztest ,critical values 
Alternative hypothesis is to be accepted.We finally conclude there is a error in the theoretial model and is shows weekly operations costs(4000) 
but actual value is higher than model predited.It is proved by using ztest to significance it.        
```
