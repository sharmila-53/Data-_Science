```python
#1. Create list of 5 integers and perform lists operations

# Create a list of 5 integers
numbers = [10, 20, 30, 40, 50]
print("Original list:", numbers)

# append
numbers.append(60)
print("After append:", numbers)

# extend
numbers.extend([70, 80])
print("After extend:", numbers)

# insert
numbers.insert(2, 25)
print("After insert:", numbers)

# remove
numbers.remove(40)
print("After remove:", numbers)

# pop
numbers.pop()
print("After pop:", numbers)

# index
print("Index of 30:", numbers.index(30))

# count
print("Count of 20:", numbers.count(20))

# sort
numbers.sort()
print("After sort:", numbers)

# reverse
numbers.reverse()
print("After reverse:", numbers)

# clear
numbers.clear()
print("After clear:", numbers)
```

```python
# 11.Creating and Exploring a Pandas DataFrame
import numpy as np
import pandas as pd

#create dataframe manually with 10rows
data = { 
    "Name": ["Sharmila", "Anu", "Ravi", "Karthik", "Priya",
             "Anu", "Suresh", "Meena", "Rahul", "Divya"],
    "Age": [22, 24, 25, 23, 26, 24, 28, 27, 25, 23],
    "City": ["Hosur", "Bangalore", "Hyderabad", "Chennai", "Mumbai",
             "Bangalore", "Delhi", "Pune", "Hyderabad", "Chennai"],
    "Salary": [25000, 30000, np.nan, 28000, 35000,
               30000, 40000, 32000, np.nan, 27000]
       }
df = pd.DataFrame(data)
df = df.reset_index(drop=True)
df.index+=1
df
```

```python
#use.info() Displays column names, data types, non-null values, and memory usage.
print("Info:\n",df.info())
```

```python
#use.describe() Displays count, mean, standard deviation, minimum, and maximum values for numerical columns.
print("\nDescribe:\n",df.describe())
```

```python
#Select 'Name' and 'city' Columns
print("\nSelecting name and city:\n",df[['Name','City']])
```

```python
#Drop 'City' Column
print("\nDrop city:\n",df.drop('City', axis=1))
```

```python
#Fill Missing Values in 'Salary' with Mean

#find mean for salary
salary_mean = df['Salary'].mean()

#fill the missing value with mean value
df['Salary'] = df['Salary'].fillna(salary_mean)

print("\nAfter Filling Missing Salary Values:\n", df)

```

```python
#Remove Duplicate Rows
df = df.drop_duplicates()
print("\nAfter Removing Duplicate Rows:\n", df)
```
